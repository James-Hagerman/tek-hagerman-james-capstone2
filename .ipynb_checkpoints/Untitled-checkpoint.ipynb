#### Sea level

```python
print(sea_level.shape)
print(sea_level.isnull().sum().sum())
sea_level.head()
```

```python
sea_level.info()
```

```python
sea_level_split1 = sea_level.drop('UHSLCFD_change_mm', axis = 1)
sea_level_split1.head()
```

```python
sea_level_split1 = sea_level_split1.dropna()
```

```python
sea_level_split1.drop(sea_level_split1[sea_level_split1['Time'] >= '1970-01-16'].index, inplace = True)
sea_level_split1.rename(columns = {'CW_change_mm': 'change_sea_level_mm'}, inplace = True)
sea_level_split1.tail()
```

```python
sea_level_split2 = sea_level.drop('CW_change_mm', axis = 1)
sea_level_split2 = sea_level_split2.dropna()
sea_level_split2.head()
sea_level_split2.rename(columns = {'UHSLCFD_change_mm': 'change_sea_level_mm'}, inplace = True)
```

```python
total_sea_level = sea_level_split1.append(sea_level_split2)
print(total_sea_level.shape)
total_sea_level.tail()
```

```python
total_sea_level['Time'] = total_sea_level['Time'].str[:4].astype(int)
```

```python
total_sea_level.describe()

total_sea_level.head(2)
```

```python
total_sea_level1 = total_sea_level.groupby(
    ['Time']
)['change_sea_level_mm'].mean().reset_index(name='change_sea_level_mm')
print(total_sea_level.shape)
total_sea_level.head(10)
```

```python
total_sea_level1['change_sea_level_mm'] = total_sea_level1['change_sea_level_mm'] + (191.58073)
total_sea_level1
```

```python
total_sea_level1['sea_level'] = total_sea_level1['change_sea_level_mm']
total_sea_level1['change_sea_level_mm'] = total_sea_level1.change_sea_level_mm.diff()
total_sea_level1.head()
```

```python
total_sea_level = total_sea_level1[1:]
```

```python
total_sea_level.head()
```

```python
total_sea_level.rename(columns = {'Time':'Year'}, inplace = True)
total_sea_level.head(15)
```

#### Temperature ðŸ”´IMPORTANTðŸ”´ Used only for joining on pollution and greenhouse gas data on 'Year' and 'Country'


```python
print(temperature.shape)
print(temperature.isnull().sum().sum())
temperature.head()
```

```python
# Visualizing null values
sns.heatmap(temperature.isnull(), cbar = False, cmap = 'viridis')
```

```python
temperature.interpolate(method ='linear', limit_direction ='backward', inplace = True)
```

```python
# Visualizing null values
sns.heatmap(temperature.isnull(), cbar = False, cmap = 'viridis')
```

```python
# Converting Temperature in Celsius to temperature in Farenheit
temperature['AverageTemperature'] = (temperature['AverageTemperature'] * 9.0/5.0) + 32.0
temperature['AverageTemperatureUncertainty'] = (temperature['AverageTemperatureUncertainty'] * 9.0/5.0) + 32.0

temperature.head()
```

```python
# Changing the date column from a string to a datetime object
temperature['dt'] = pd.to_datetime(temperature['dt'])
temperature.info()
```

```python
# Getting the average temperature every year for each country 
temperature = temperature.groupby(
    [temperature['dt'].dt.strftime('%Y'), 'Country']
)['AverageTemperature'].mean().reset_index(name='AverageTemperature')
```

```python
temperature = temperature.sort_values(by=['Country', 'dt'], ascending=True)
```

```python
temperature.rename(columns = {'dt':'Year'}, inplace = True)
```

```python
temperature.tail()
```

```python
gh_total = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/greenhouse/gh_total.csv')
gh_carbon_dioxide = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/greenhouse/gh_carbon_dioxide.csv')
gh_methane = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/greenhouse/gh_methane.csv')
gh_nitrous_oxide = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/greenhouse/gh_nitrous_oxide.csv')
gh_hydrofluorocarbons = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/greenhouse/gh_hydrofluorocarbons.csv')
gh_perfluorocarbons = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/greenhouse/gh_perfluorocarbons.csv')
gh_sulphur_hexaflouride = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/greenhouse/gh_sulphur_hexafluoride.csv')
```

### Pollutants

```python
p_non_methane_VOCs = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/pollutants/p_non_methane_VOCs.csv')
p_carbon_monoxide = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/pollutants/p_carbon_monoxide.csv')
p_nitrogen_oxides = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/pollutants/p_nitrogen_oxides.csv')
p_sulphur_oxides = pd.read_csv('hagerman_james_capstone2/Data/Air_pollutants/Merging datasets/pollutants/p_sulphur_oxides.csv')
```

#### Cleaning


```python
# Cleaning p_non_methane_vocs
p_non_methane_VOCs = clean_df(p_non_methane_VOCs)
p_non_methane_VOCs.rename(columns = {'Value': 'non_methane_VOCs_value'}, inplace = True)
p_non_methane_VOCs.head(1)
```

```python
# Cleaning p_carbon_monoxide
p_carbon_monoxide = clean_df( p_carbon_monoxide)
p_carbon_monoxide.rename(columns = {'Value': 'carbon_monoxide_value'}, inplace = True)
p_carbon_monoxide.head(1)
```

```python
# Cleaning p_nitrogen_oxides
p_nitrogen_oxides = clean_df(p_nitrogen_oxides)
p_nitrogen_oxides.rename(columns = {'Value': 'nitrogen_oxide_value'}, inplace = True)
p_nitrogen_oxides.head(1)
```

```python
# Cleaning p_sulphur_oxides
p_sulphur_oxides = clean_df(p_sulphur_oxides)
p_sulphur_oxides.rename(columns = {'Value': 'sulphur_oxide_value'}, inplace = True)
p_sulphur_oxides.head(1)
```

#### Aggregating

```python
# Joining p_carbon_monoxide onto p_non_methane_vocs
p_total = pd.merge(p_non_methane_VOCs, p_carbon_monoxide, how = 'left', on = ['Country', 'Year'])
print(p_total.shape)
p_total.head(1)
```

```python
# Joining p_nitrogen_oxides onto p_total
p_total = pd.merge(p_total, p_nitrogen_oxides, how = 'inner', on = ['Country', 'Year'])
print(p_total.shape)
p_total.head(1)
```

```python
# Joining p_sulphur_oxides onto p_total
p_total = pd.merge(p_total, p_sulphur_oxides, how = 'inner', on = ['Country', 'Year'])
print(p_total.shape)
p_total.head(1)
```

```python
# Creating a column that is the total pollution 
sum_column = p_total['non_methane_VOCs_value'] + p_total['carbon_monoxide_value'] + p_total['nitrogen_oxide_value'] + p_total['sulphur_oxide_value'] 
p_total['total_pollution'] = sum_column
```

```python
# Visualizing null values
sns.heatmap(p_total.isnull(), cbar = False, cmap = 'viridis')
```

### Death 

```python
deaths_ambient_particulate = pd.read_csv('hagerman_james_capstone2/Data/Death/merging/deaths_p_ambient_particulate_matter.csv')
deaths_air_pol_solid_fuels = pd.read_csv('hagerman_james_capstone2/Data/Death/merging/deaths_p_air_pol_solid_fuels.csv')
deaths_ambient_ozone = pd.read_csv('hagerman_james_capstone2/Data/Death/merging/deaths_p_abient_ozone.csv')
deaths_high_temp = pd.read_csv('hagerman_james_capstone2/Data/Death/merging/deaths_t_high_temp.csv')
deaths_low_temp = pd.read_csv('hagerman_james_capstone2/Data/Death/merging/deaths_t_low_temp.csv')
```

```python
# Function to remove useless columns from datasets
def clean_df(df):
    df.drop(['COU', 'VAR', 'YEA', 'SEX', 'Sex', 'Age', 'AGE', 'RISK', 'Unit Code', 'Risk', 'Unit', 'PowerCode Code', 'PowerCode', 
             'Reference Period Code', 'Reference Period', 'Flag Codes', 'Flags', 'Variable', 'Unit'], axis = 1, inplace = True) 
    return df
```

#### Cleaning

```python
# Cleaning deaths_ambient_particulate 
deaths_ambient_particulate = clean_df(deaths_ambient_particulate)
deaths_ambient_particulate.rename(columns = {'Value': 'ambient_particulate_deaths'}, inplace = True)
deaths_ambient_particulate.head()
```

```python
# Cleaning deaths_air_pol_solid_fuels 
deaths_air_pol_solid_fuels = clean_df(deaths_air_pol_solid_fuels)
deaths_air_pol_solid_fuels.rename(columns = {'Value': 'air_pol_solid_fuel_deaths'}, inplace = True)
deaths_air_pol_solid_fuels.head()
```

```python
# Cleaning deaths_ambient_ozone 
deaths_ambient_ozone = clean_df(deaths_ambient_ozone)
deaths_ambient_ozone.rename(columns = {'Value': 'ambient_ozone_deaths'}, inplace = True)
deaths_ambient_ozone.head()
```

```python
# Cleaning deaths_high_temp 
deaths_high_temp = clean_df(deaths_high_temp)
deaths_high_temp.rename(columns = {'Value': 'high_temp_deaths'}, inplace = True)
deaths_high_temp.head()
```

```python
# Cleaning deaths_low_temp 
deaths_low_temp = clean_df(deaths_low_temp)
deaths_low_temp.rename(columns = {'Value': 'low_temp_deaths'}, inplace = True)
deaths_low_temp.head()
```
